```

ou
```

## Alguns exercícios interessantes
Gere um dataset indicando quais gols foram marcados no primeiro tempo.

Por que a variável `minuto` é de texto? É porque os gols nos acréscimos são notados pelo texto `90+n` ou `45+n`.

````python
gols.columns
````

````python
gols.dtypes
````

````python
gols.tail(20)
````

````python
"a" > "b"
````

````python
gols[gols.minuto > '90']   # Podemos usar <, >, <=, >= com texto, comparação alfabética
````

````python
'5' < '45'
````

Infelizmente, só usar `"minuto" <= 45` não dá certo, pois exclui os gols entre os 5 e os 9 minutos.

````python
"9" <= "45"
````

````python
9 < 45
````

Para consertar, incluímos os minutos com tamanho 1 no primeiro tempo. Para passar a função `len` em uma coluna, use `df.coluna.str.len()`.

````python
gols[gols.minuto.str.len() == 1]
````

Não podemos esquecer os gols nos acréscimos do primeiro tempo:

````python
gols[(gols.minuto >= "45+") & (gols.minuto < "46")]
````

Agora estamos prontos para separar os dois tempos!

````python
gols["primeiro_tempo"] = ((gols.minuto <= "45") |
    ((gols.minuto >= "45+") & (gols.minuto < "46")) |
    (gols.minuto.str.len() == 1))

gols.head(10)
````

## Alguns gráficos
O `pandas` já tem alguns gráficos (do `matplotlib`) embutidos no `DataFrame` (são os que a IA aqui sugere).

Basta usar `df.plot(kind="que gráfico?", title="Título do gráfico")`

Os tipos de gráfico são:
* Histograma: `kind="hist"`
* Linha (padrão): `kind="line"`
* Colunas: `kind="bar"`
* Barras: `kind="barh"`
* Boxplot: `kind="box"`
* Pizza: `kind="pie"`
* Dispersão: `kind="scatter"`

A documentação tem bem mais detalhes:
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html

Vamos processar o tempo do gol no primeiro tempo para fazer um histograma.

````python
primeiro_tempo = gols[gols.primeiro_tempo]

primeiro_tempo.head()
````

Para resolver o problema, usamos a função `eval(texto)`, que faz a conta dentro do texto. Por exemplo, `eval("2+2")` dá 4.

````python
eval("90+2")
````

````python
# A função eval faz a conta dentro de uma string. Por exemplo, eval('90+2') dá 92
primeiro_tempo["minuto_num"] = [eval(x) for x in primeiro_tempo["minuto"]]

# Dá warning, mas dá certo
````

````python
primeiro_tempo.head()
````

````python
primeiro_tempo["minuto_num"].plot(kind="hist", title="Gols por minuto")
````
