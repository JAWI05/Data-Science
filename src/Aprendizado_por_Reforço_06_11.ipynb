```python
env.reset()
acabou = False

print(env.render())
while not acabou:
  acao = int(input("Passo (0=esq, 1=baixo, 2=dir, 3=cima): "))
  output.clear()
  estado, recomp, acabou, _, _ = env.step(acao)
  print(f"Estado: {estado}, Recompensa: {recomp}, Acabou: {acabou}")
  _, _, acabou, _, _ = env.step(acao)
  print(env.render())
```
