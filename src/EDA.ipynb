# A biblioteca `seaborn`
O `seaborn` é uma interface do `matplotlib`, que serve para visualizar dados e fazer gráficos.

Um tutorial bem bacana está em
https://seaborn.pydata.org/tutorial.html

```python
import seaborn as sns
```

```python
sns.histplot(data=df, x='sepal length')
```

```python
sns.displot(data=df, x='sepal width', col='class', bins=10, hue='class')
```

```python
sns.relplot(data=df, x='sepal length', y='sepal width', hue='class')
```

```python
sns.relplot(data=df, x='petal length', y='petal width', hue='class')
```

```python
sns.jointplot(data=df, x='sepal width', y='petal length', hue='class')
```

```python
sns.pairplot(df, hue='class')
```

```python
sns.boxplot(data=df, x='class', y='sepal width', hue='class')
```
