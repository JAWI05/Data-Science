# Classes
Em python podemos criar classes com a keyword class


self - objeto da classe

```python
class DataScience2:
  def __init__(self, aluno, notaP1, notaP2):
    self.name = aluno
    self.notaP1 = notaP1
    self.notaP2 = notaP2

  def media(self):
    return (self.notaP1 + self.notaP2) / 2
```

```python
joao = DataScience2("João", 6, 6.5)
```

```python
joao.media()
```

```python
class knn:
  def __init__(self, k):
    self.k = k

  def distancia(self, u, v):
    return np.sqrt(np.sum((u - v)**2))

  def resposta(self, labels):
    contagem = Counter(labels)
    vencedor, cont_vencedor = contagem.most_common(1)[0]

    n_vencedores = len([i for i in contagem.values() if i == vencedor])
    if n_vencedores == 1:
      return vencedor
    else:
      return self.resposta(labels[:-1])
```

```python
u = np.array([0,0])
v = np.array([3,4])
```

```python
teste = knn(5)
teste.distancia(u,v)
```
